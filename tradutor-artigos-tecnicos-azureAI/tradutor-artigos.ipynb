```python
import requests
from bs4 import BeautifulSoup


def extrair_texto_url(url):
    try:
        # Fazer a requisição para a URL
        response = requests.get(url)

        if response.status_code == 200:
            # Parsear o HTML da página
            soup = BeautifulSoup(response.text, 'html.parser')

            # Remover scripts e estilos para evitar textos desnecessários
            for script_or_style in soup(['script', 'style']):
                script_or_style.decompose()

            # Extrair o texto da página
            texto = soup.get_text(separator=' ')

            # Limpar e formatar o texto extraído
            linhas = (line.strip() for line in texto.splitlines())
            parts = (phrase.strip()
                     for line in linhas for phrase in line.split(" "))
            texto_limpo = '\n'.join(part for part in parts if part)

            return texto_limpo
        else:
            print(
                f"Falha ao buscar a URL. Código de status: {response.status_code}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Erro ao acessar a URL: {e}")
        return None


# Exemplo de uso
texto = extrair_texto_url('https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo')
if texto:
    print(texto)
```

```python
from langchain_openai.chat_models.azure import AzureChatOpenAI

client = AzureChatOpenAI(
    azure_endpoint="https://meu-servico-ai.openai.azure.com/",
    api_key= "meu_key",
    api_version= "2024-02-15-preview",
    deployment_name= "gpt-4o-mini",
    max_retries=0
)

def traduzir_artigo(text, lang):
    messages = [
        ("system", "Você atua como tradutor de textos"),
        ("user", f"Traduza o {text} para o idioma {lang} e responda em markdown")
    ]
    
    response = client.invoke(messages)
    print(response.content)
    return response.content

traduzir_artigo("Let's see if the deployment was succeeded.", "portugues")
```

```python
url = 'https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo'
text = extrair_texto_url(url)
article = traduzir_artigo(text, "pt-br")

print(article)
```
